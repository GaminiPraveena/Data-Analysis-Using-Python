Extracting a table from a webpage

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url1="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
page1=requests.get(url1)
code1=BeautifulSoup(page1.text,'html')
print(code1)
```

```python
t1=code1.find('table')
print(t1)
```

```python
th=t1.find_all('th')
print(th)
```

```python
headings=[i.text for i in th]
print(headings)
```

```python
import pandas as pd
mydf=pd.DataFrame(columns=headings) #mydf=pd.dataframe(columns=headings)(text)
print(mydf)
```

```python
rows=t1.find_all('tr')
print(rows)
```

```python
data=t1.find_all('td')
print(data)
```

```python
data_text1=[i.text for i in data]
print(data_text1)
```

```python
start=0
stop=4
ind=0
for i in data_text1:
  if stop<=40:           #len(data_text)
    mydf.loc[ind]=data_text1[start:stop]
    start=start+4
    stop=stop+4
    ind=ind+1
print(mydf)
```
