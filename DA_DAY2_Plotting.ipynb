```python
#there are 5 subjects psychology,nanotechnology,medicine,onthologyand anthropology 2 students have applied to do masters in any of these 5 fields
# with a line plot show in which field they can excel the cut off for each subject is 70
#student1[67,45,73,71,57]
#student2[43,67,68,53,85]
plt.title("PLOT")
sub=["Psy","Nano","Med","Onc","Anthro"]
student1=[67,45,73,71,57]
student2=[43,67,68,53,85]
cutoff=[70,70,70,70,70]

#student1
plt.scatter(sub,student1,label='student1',color='r',marker="P")
#student2
plt.scatter(sub,student2,label='student2',color='b',marker="H")
#cutoff
plt.plot(sub,cutoff,label="cutoff",color='g',linestyle=":",linewidth=4)
plt.xlabel('subjects')
plt.ylabel("marks")
plt.legend()
```
