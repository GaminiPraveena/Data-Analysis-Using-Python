```python
# creating set
s={12.3,'air','hi','hello',32}  #no index
print(s)
#creating using set function
s1=set((12,12.12,'veena','gpt'))
print(s1)
print('veena' in s1)
print("s length=",len(s),"s1 length=",len(s1))
```

```python
#printing elements of a set
for i in s:
  print(i)
```

```python
#adding new element into the set s1
s1.add("added")
print(s1)
```

```python
#updating list into a set
l=[1,2,3]
s.update(l)
print(s)
```

```python
#removing is done in 3 ways
#1.remove(element):raises error if the element is not present
#2.discard(element):dont raises error if the element is not present
#3.pop():removes random element
s3={1,2,3,"h","a","b","e"}
s3.remove(1)
s3.discard(2)
s3.pop()
print(s3)
```

```python
#clearing the set
s.clear()
print(s)
```

```python
#union
a={1,2,3}
b={5,6,78,45}
print(a.union(b))
```

```python
#intersection-returns a set
#intersection_update-dont returns a set
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.intersection(y)

print(z)
```

```python
#symmetric_difference-returns a set
#symmetric_differnece_update-dont returns a set
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.symmetric_difference(y)

print(z)
```

```python
#sort changes the original set sorted
a={3,1,6,2}
a.sort()
print(a)
```

```python
#creating a tuple -allows duplicate values,have indexing,unchangeable
t=(12,'ear',12.3,'eye',32,12) #t=tuple((12,'ear',12.3,'eye',32,12))
print(t)
```

```python
#accessing elements in tuple
print(t[3])
print(t[-2])
```

```python
#changing elements in a tuple by converting it into a list
l1=list(t)
l1[2]=100
t=tuple(l1)
print(t)
```

```python
#joining two tuples
t1=(1,2,3)
t2=("fair","styles",12)
print(t1+t2)
```
